```python
from gsdmm.mgp import MovieGroupProcess
```

```python
texts=[]
with open('../cleaned_english_tweets_newData.txt', 'r', encoding='utf-8') as filehandle:
    for line in filehandle:
        # remove linebreak which is the last character of the string
        text = line[:-1]
        print(text)

        # add item to the list
        texts.append(text)
        
texts
```

```python
def setUp():
        numpy.random.seed(47)

def tearDown():
    numpy.random.seed(None)

def compute_V(texts):
    V = set()
    for text in texts:
        for word in text:
            V.add(word)
    return len(V)
```

```python
texts = [text.split() for text in texts]
V = compute_V(texts)
mgp = MovieGroupProcess(K=2, n_iters=50, alpha=0.2, beta=0.01)
y = mgp.fit(texts, V)
```

```python
print(len(y))
print(y)
print(len(texts))
```

```python
topics_dictionary = {}
for index, tweet in enumerate(texts):
    if y[index] in topics_dictionary:
        topics_dictionary[y[index]].append(tweet)
    else:
        topics_dictionary[y[index]] = [tweet]
```

```python
topics_dictionary
```

```python
import numpy as np 
import operator

doc_count = np.array(mgp.cluster_doc_count)
print('Number of documents per topics :', doc_count)
print('*'*20)

# Topics sorted by document inside
top_index = doc_count.argsort()[-10:][::-1]
print('Most important clusters (by number of docs inside):', top_index)
print('*'*20)

for i, topic in enumerate(mgp.cluster_word_distribution):
    sorted_topics = sorted(topic.items(), key=operator.itemgetter(1),
                            reverse = True)
    print('\n Topic: {}'.format(i))
    print(sorted_topics[:10])
```

```python
from numpy.random import multinomial
from numpy import log, exp
from numpy import argmax

def score(doc, V):    
    alpha, beta, K, D = 0.2, 0.01, 2, 5305
    m_z, n_z, n_z_w = [0 for _ in range(K)], [0 for _ in range(K)], [{} for i in range(K)]

    p = [0 for _ in range(K)]

    #  We break the formula into the following pieces
    #  p = N1*N2/(D1*D2) = exp(lN1 - lD1 + lN2 - lD2)
    #  lN1 = log(m_z[z] + alpha)
    #  lN2 = log(D - 1 + K*alpha)
    #  lN2 = log(product(n_z_w[w] + beta)) = sum(log(n_z_w[w] + beta))
    #  lD2 = log(product(n_z[d] + V*beta + i -1)) = sum(log(n_z[d] + V*beta + i -1))

    lD1 = log(D - 1 + K * alpha)
    doc_size = len(doc)
    for label in range(K):
        lN1 = log(m_z[label] + alpha)
        lN2 = 0
        lD2 = 0
        for word in doc:
            lN2 += log(n_z_w[label].get(word, 0) + beta)
        for j in range(1, doc_size +1):
            lD2 += log(n_z[label] + V * beta + j - 1)
        print(lN1 - lD1 + lN2 - lD2)
        print(lN1, lD1, lN2, lD2)
        p[label] = exp(lN1 - lD1 + lN2 - lD2)
    print(p)

    # normalize the probability vector
    pnorm = sum(p)
    pnorm = pnorm if pnorm>0 else 1
    return [pp/pnorm for pp in p]
```
