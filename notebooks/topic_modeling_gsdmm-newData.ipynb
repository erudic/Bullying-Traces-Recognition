```python
from gsdmm.mgp import MovieGroupProcess
```

```python
texts=[]
with open('../cleaned_english_tweets_newData.txt', 'r', encoding='utf-8') as filehandle:
    for line in filehandle:
        # remove linebreak which is the last character of the string
        text = line[:-1]
        print(text)

        # add item to the list
        texts.append(text)
        
print(texts)
texts = [text.split() for text in texts]
texts
```

```python
def setUp():
        numpy.random.seed(47)

def tearDown():
    numpy.random.seed(None)

def compute_V(texts):
    V = set()
    for text in texts:
        for word in text:
            V.add(word)
    return len(V)
```

```python
texts = [text.split() for text in texts]
V = compute_V(texts)
mgp = MovieGroupProcess(K=2, n_iters=50, alpha=0.2, beta=0.01)
y = mgp.fit(texts, V)
```

```python
print(len(y))
print(y)
print(len(texts))
```

```python
topics_dictionary = {}
for index, tweet in enumerate(texts):
    if y[index] in topics_dictionary:
        topics_dictionary[y[index]].append(tweet)
    else:
        topics_dictionary[y[index]] = [tweet]
```
