```python
from gsdmm.mgp import MovieGroupProcess
mgp = MovieGroupProcess(K=8, alpha=0.1, beta=0.1, n_iters=30)
```

```python
texts=[]
with open('../cleaned_english_tweets_newData.txt', 'r', encoding='utf-8') as filehandle:
    for line in filehandle:
        # remove linebreak which is the last character of the string
        text = line[:-1]
        print(text)

        # add item to the list
        texts.append(text)
        
texts
```

```python
def setUp():
        numpy.random.seed(47)

def tearDown():
    numpy.random.seed(None)

def compute_V(texts):
    V = set()
    for text in texts:
        for word in text:
            V.add(word)
    return len(V)
```

```python
texts = [text.split() for text in texts]
V = compute_V(texts)
mgp = MovieGroupProcess(K=2, n_iters=1000, alpha=0.2, beta=0.01)
y = mgp.fit(texts, V)
```

```python
print(len(y))
print(y)
print(len(texts))
```

```python
topics_dictionary = {}
for index, tweet in enumerate(texts):
    if y[index] in topics_dictionary:
        topics_dictionary[y[index]].append(tweet)
    else:
        topics_dictionary[y[index]] = [tweet]
```

```python
topics_dictionary
```

```python
import numpy as np 
import operator

doc_count = np.array(mgp.cluster_doc_count)
print('Number of documents per topics :', doc_count)
print('*'*20)

# Topics sorted by document inside
top_index = doc_count.argsort()[-10:][::-1]
print('Most important clusters (by number of docs inside):', top_index)
print('*'*20)

for i, topic in enumerate(mgp.cluster_word_distribution):
    sorted_topics = sorted(topic.items(), key=operator.itemgetter(1),
                            reverse = True)
    print('\n Topic: {}'.format(i))
    print(sorted_topics[:10])
```
